```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image_dataset_from_directory
import os

# --- 1. SETUP & PATHS ---
BASE_DIR = r"C:\Users\achyu\OneDrive\Documents\Crop-Advisory-System"
DATASET_DIR = os.path.join(BASE_DIR, "dataset", "PlantVillage", "color")
MODEL_SAVE_PATH = os.path.join(BASE_DIR, "models", "plant_disease_cnn.h5")

BATCH_SIZE = 32
IMG_SIZE = (224, 224) 
EPOCHS = 10

# --- 2. LOAD DATASET ---
print("⏳ Step 1: Scanning hard drive for images (This takes 5-10 minutes due to OneDrive...)")
train_dataset = image_dataset_from_directory(
    DATASET_DIR,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_dataset = image_dataset_from_directory(
    DATASET_DIR,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

class_names = train_dataset.class_names
print(f"\n✅ Success! Detected {len(class_names)} total crop disease classes.")

AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.prefetch(buffer_size=AUTOTUNE)

# --- 3. BUILD AI ARCHITECTURE ---
print("\n⏳ Step 2: Building Neural Network in RAM...")
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False 

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.2)(x) 
outputs = Dense(len(class_names), activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=outputs)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# --- 4. THE VISIBLE PROGRESS TRACKER ---
class PrintProgressCallback(tf.keras.callbacks.Callback):
    def on_epoch_begin(self, epoch, logs=None):
        print(f"\n🚀 --- STARTING EPOCH {epoch + 1}/{self.params['epochs']} ---")

    def on_train_batch_end(self, batch, logs=None):
        # This forces a print statement every 50 batches
        if batch > 0 and batch % 50 == 0:
            total = self.params['steps']
            accuracy = logs.get('accuracy', 0) * 100
            print(f"  ⚙️ Processing Batch {batch}/{total} | Current Accuracy: {accuracy:.1f}%")

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    MODEL_SAVE_PATH, 
    monitor='val_accuracy', 
    save_best_only=True, 
    verbose=0 
)

# --- 5. EXECUTE TRAINING ---
print("\n⏳ Step 3: Loading images into memory for math (Please wait a few more minutes...)")
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=EPOCHS,
    verbose=2, # Turns off invisible animations
    callbacks=[checkpoint, PrintProgressCallback()] # Adds our custom tracker
)

print(f"\n✅ TRAINING COMPLETE! Model saved successfully to: {MODEL_SAVE_PATH}")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
import pickle
import os
from datetime import datetime, timedelta

# --- 1. CONFIGURATION & PATHS ---
BASE_DIR = r"C:\Users\achyu\OneDrive\Documents\Crop-Advisory-System"
DATA_PATH = os.path.join(BASE_DIR, "dataset", "apmc_historical_prices.csv")
MODEL_SAVE_PATH = os.path.join(BASE_DIR, "models", "apmc_price_lstm.h5")
SCALER_SAVE_PATH = os.path.join(BASE_DIR, "models", "price_scaler.pkl")

LOOKBACK_DAYS = 30
EPOCHS = 15
BATCH_SIZE = 32

# --- 2. FORCE GENERATE CLEAN DATASET ---
print("🧹 Cleaning old files and generating fresh synthetic APMC dataset...")
start_date = datetime(2023, 1, 1)
dates = [start_date + timedelta(days=i) for i in range(1095)]

np.random.seed(42)
prices = []
for i, date in enumerate(dates):
    trend = i * 0.5
    # FIX APPLIED HERE: changed yday to tm_yday
    seasonal_spike = 800 * np.sin(2 * np.pi * date.timetuple().tm_yday / 365.25)
    weekly_noise = 150 * np.sin(2 * np.pi * i / 7)
    random_shock = np.random.normal(0, 100)
    modal_price = max(800, 2000 + trend + seasonal_spike + weekly_noise + random_shock)
    prices.append(int(modal_price))

# Overwrite the corrupted file
df_gen = pd.DataFrame({"Date": dates, "Commodity": "Tomato", "APMC": "Anantapur", "Modal_Price": prices})
os.makedirs(os.path.dirname(DATA_PATH), exist_ok=True)
df_gen.to_csv(DATA_PATH, index=False)
print("✅ Clean dataset generated and saved successfully!")

# --- 3. LOAD & PREPROCESS DATA ---
print("📊 Loading dataset into Pandas...")
df = pd.read_csv(DATA_PATH)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

price_data = df['Modal_Price'].values.reshape(-1, 1)

# Scale data for the LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(price_data)

os.makedirs(os.path.dirname(SCALER_SAVE_PATH), exist_ok=True)
with open(SCALER_SAVE_PATH, 'wb') as f:
    pickle.dump(scaler, f)
print("✅ Price Scaler saved successfully.")

# --- 4. CREATE WINDOWED SEQUENCES ---
X, y = [], []
for i in range(LOOKBACK_DAYS, len(scaled_data)):
    X.append(scaled_data[i-LOOKBACK_DAYS:i, 0])
    y.append(scaled_data[i, 0])

X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

split = int(len(X) * 0.8)
X_train, X_val = X[:split], X[split:]
y_train, y_val = y[:split], y[split:]

print(f"Training shape: {X_train.shape}, Validation shape: {X_val.shape}")

# --- 5. BUILD THE LSTM ARCHITECTURE ---
print("\n🧠 Building LSTM Network...")
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(LOOKBACK_DAYS, 1)),
    Dropout(0.2),
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),
    Dense(units=25),
    Dense(units=1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

# --- 6. TRAIN AND SAVE ---
print("\n🚀 Starting LSTM Training...")
model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=1
)

model.save(MODEL_SAVE_PATH)
print(f"\n✅ LSTM Price Forecasting Model saved to: {MODEL_SAVE_PATH}")
```
